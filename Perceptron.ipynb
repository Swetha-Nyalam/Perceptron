```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#defining some random data set
X = np.array([[10,6],[8,5],[-3,-3],[-4,1],[0,6]])
Y = np.array([1,1,-1,-1, -1])
#dividing into two classes for plotting
x_class0 = X[Y==1]
x_class1 = X[Y == -1]
```

```python
#defining epochs and learning_rate
epochs = 6
lr = 0.01
```

```python
class Perceptron:
    def __init__(self, lr:float, epochs : int):
        self.lr = lr
        self.epochs = epochs
        
    def predict(self,x):
        pred_val = np.dot(x,self.weights)+self.bias
        return np.where(pred_val >= 0.0 , 1, -1)
           
    def fit(self,X,Y):
        m,n = X.shape
        self.weights = np.zeros(n)
        self.bias = 0
        for _ in range(epochs):
            for i in range(m):
                y_pred = self.predict(X[i])
                error = self.lr*(Y[i]-y_pred)
                self.weights += error*X[i]
                self.bias+= error 
        print(self.weights)
        print(self.bias)
        return self
```

```python
perceptron = Perceptron(lr,epochs)
perceptron.fit(X, Y)
perceptron.predict(np.array([[-3,-3],[10,6],[10,6],[8,5],[-3,-3],[-4,1],[0,6]]))
```
